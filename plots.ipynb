```python
import matplotlib.pyplot as plt
import numpy as np
import run_strands

test = run_strands.TestClass()
test.test()
```

### Plot Creation
- Plot 1: Time taken to find solution for each algorithm
- Plot 2: Total number of recursions per algorithm
- Plot 3: Total number of incorrect guesses per algorithm

```python
labels = ["Word Rules", "Greedy", "Dictionary", "Custom"]
values = [test.wordy_time, test.greedy_time, test.dict_time, test.custom_time]
fig = plt.figure(figsize = (10, 5))
plt.barh(labels, values, color='#f8cd05ff')
plt.ylabel("Algorithm")
plt.xlabel("Seconds elapsed")
plt.title("Time elapsed per search algorithm")
plt.show()
```

```python
values2 = [test.wordy_solver.recursion_count, test.dfs_solver.recursion_count, test.dict_solver.recursion_count, test.custom_solver.recursion_count]
fig2 = plt.figure(figsize = (10, 5))
plt.barh(labels, values2, color='#c0ddd9ff')
plt.ylabel("Algorithm")
plt.xlabel("Times recursed")
plt.title("Recursion count per search algorithm")
plt.show()
```

```python
values3 = [test.wordy_solver.incorrect_guesses, test.dfs_solver.incorrect_guesses, test.dict_solver.incorrect_guesses, test.custom_solver.incorrect_guesses]
print(test.dict_solver.incorrect_guesses)
fig3 = plt.figure(figsize = (10, 5))
plt.barh(labels, values3, color='#aedfeeff')
plt.ylabel("Algorithm")
plt.xlabel("Incorrect Guesses")
plt.title("Guess correctness per search algorithm")
plt.show()
```

```python
short_labels = ["Dictionary", "Custom"]
short_guesses = [test.dict_solver.incorrect_guesses, test.custom_solver.incorrect_guesses]
short_recursions = [test.dict_solver.recursion_count, test.custom_solver.recursion_count]
n = np.arange(len(short_labels))
width = 0.25
# fig, ax = plt.subplots()
plt.bar(n - 0.2, short_guesses, 0.4, label = 'Incorrect Guesses', color='#f8cd05ff')
plt.bar(n + 0.2, short_recursions, 0.4, label = 'Recursions', color='#aedfeeff')
plt.xticks(n, short_labels) 
plt.xlabel("Algorithms")
plt.ylabel("Efficiency")
plt.title("Efficiency of Dictionary and Custom algorithms")
plt.legend()
plt.show()
```
