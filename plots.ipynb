```python
import main
import boards
import matplotlib.pyplot as plt
import numpy as np
import run_strands

test_small = run_strands.TestClass(boards.small_board, boards.small_solution, boards.small_width, boards.small_height)
test_med = run_strands.TestClass(boards.stork_board, boards.stork_solution, boards.stork_width, boards.stork_height)
test_large = run_strands.TestClass(boards.cs_board, boards.cs_solution, boards.cs_width, boards.cs_height)
test_small.test()
test_med.test()
test_large.test()
```

### Plot Creation
- Plot 1: Time taken to find solution for each algorithm
- Plot 2: Total number of recursions per algorithm
- Plot 3: Total number of incorrect guesses per algorithm

```python
labels = ["Word Rules", "Greedy", "Custom"]
word_rules_vals = [test_small.wordy_time, test_med.wordy_time, test_large.wordy_time]
greedy_vals = [test_small.greedy_time, test_med.greedy_time, test_large.greedy_time]
custom_vals = [test_small.custom_time, test_med.custom_time, test_large.custom_time]
x = [1, 2, 3]

print(word_rules_vals)
print(greedy_vals)
print(custom_vals)

plt.plot(x, word_rules_vals, label="Word Rules", marker='o')
plt.plot(x, greedy_vals, label="Greedy", marker='o', color='orange')
plt.plot(x, custom_vals, label="Custom", marker='o', color='green')
plt.xticks([1, 2, 3], ['Small Board', 'Medium Board', 'Large Board'], rotation=20)
plt.legend()
plt.ylabel("Seconds elapsed")
plt.title("Increase in complexity across three board sizes")
plt.show
```

```python
labels = ["Word Rules", "Greedy", "Dictionary", "Custom"]
values = [test_med.wordy_time, test_med.greedy_time, test_med.dict_time, test_med.custom_time]
fig = plt.figure(figsize = (10, 5))
plt.barh(labels, values, color='#f8cd05ff')
plt.ylabel("Algorithm")
plt.xlabel("Seconds elapsed")
plt.title("Time elapsed per search algorithm")
plt.show()
```

```python
values2 = [test_med.wordy_solver.recursion_count, test_med.dfs_solver.recursion_count, test_med.dict_solver.recursion_count, test_med.custom_solver.recursion_count]
fig2 = plt.figure(figsize = (10, 5))
plt.barh(labels, values2, color='#c0ddd9ff')
plt.ylabel("Algorithm")
plt.xlabel("Times recursed")
plt.title("Recursion count per search algorithm")
plt.show()
```

```python
values3 = [test_med.wordy_solver.incorrect_guesses, test_med.dfs_solver.incorrect_guesses, test_med.dict_solver.incorrect_guesses, test_med.custom_solver.incorrect_guesses]
fig3 = plt.figure(figsize = (10, 5))
plt.barh(labels, values3, color='#aedfeeff')
plt.ylabel("Algorithm")
plt.xlabel("Incorrect Guesses")
plt.title("Guess correctness per search algorithm")
plt.show()
```

```python
short_labels = ["Dictionary", "Custom"]
short_guesses = [test_med.dict_solver.incorrect_guesses, test_med.custom_solver.incorrect_guesses]
short_recursions = [test_med.dict_solver.recursion_count, test_med.custom_solver.recursion_count]
n = np.arange(len(short_labels))
width = 0.25
# fig, ax = plt.subplots()
plt.bar(n - 0.2, short_guesses, 0.4, label = 'Incorrect Guesses', color='#f8cd05ff')
plt.bar(n + 0.2, short_recursions, 0.4, label = 'Recursions', color='#aedfeeff')
plt.xticks(n, short_labels) 
plt.xlabel("Algorithms")
plt.ylabel("Efficiency")
plt.title("Efficiency of Dictionary and Custom algorithms")
plt.legend()
plt.show()
```
